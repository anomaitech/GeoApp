```python


# Import libraries
import geopandas as gpd
import matplotlib.pyplot as plt
import mapclassify
import libpysal

# Load the 'South' dataset from libpysal (cells [5] and [6])
# This dataset contains US Southern county homicide data
_ = libpysal.examples.load_example('South')
pth = libpysal.examples.get_path('south.shp')
df = gpd.read_file(pth)

# Default legend formatting with Quantiles (cell [7])
# Plot homicide rates (HR60) with 4 quantile classes
plt.figure(figsize=(10, 6))
ax = df.plot(
    column='HR60',  # Homicide rate per 100,000 in 1960
    scheme='QUANTILES',  # Classify into 4 quantiles
    k=4,  # Number of classes
    cmap='BuPu',  # Blue-Purple color scheme
    legend=True,
    legend_kwds={'loc': 'center left', 'bbox_to_anchor': (1, 0.5)}  # Position legend
)
plt.title('Homicide Rates (HR60) - Quantiles')
plt.show()

# Override numerical format (cell [13])
# Use a custom format for 4 decimal places
plt.figure(figsize=(10, 6))
ax = df.plot(
    column='HR60',
    scheme='QUANTILES',
    k=4,
    cmap='BuPu',
    legend=True,
    legend_kwds={'loc': 'center left', 'bbox_to_anchor': (1, 0.5), 'fmt': '{:.4f}'}
)
plt.title('Homicide Rates (HR60) - Quantiles with Custom Format')
plt.show()

# BoxPlot scheme with custom format (cell [14])
# Use BoxPlot classification to detect outliers
plt.figure(figsize=(10, 6))
ax = df.plot(
    column='HR60',
    scheme='BoxPlot',
    cmap='BuPu',
    legend=True,
    legend_kwds={'loc': 'center left', 'bbox_to_anchor': (1, 0.5), 'fmt': '{:.0f}'}
)
plt.title('Homicide Rates (HR60) - BoxPlot')
plt.show()

# Show interval brackets (cell [17])
# Add brackets to show open/closed intervals
plt.figure(figsize=(10, 6))
ax = df.plot(
    column='HR60',
    scheme='BoxPlot',
    cmap='BuPu',
    legend=True,
    legend_kwds={'loc': 'center left', 'bbox_to_anchor': (1, 0.5), 'interval': True}
)
plt.title('Homicide Rates (HR60) - BoxPlot with Intervals')
plt.show()
```
